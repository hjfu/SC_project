```
print(sc_df_0.describe())
sc_df_0 = sc_df_0.replace(r'', np.nan)
```

```
# get the missing value counts in percentages

sc_nan_df = sc_df_0.apply(lambda x: (x.isnull().sum()*100/len(x)))
print(sc_nan_df) 
```

```
#get outlier values for each variable by constructing z score

sc_z_df = sc_df_0.apply(lambda x: (x - x.mean())/x.std())
sc_outlier_df = sc_z_df.abs() > 3
# print(sc_outlier_df)
# sc_df_0[sc_outlier_df.any(axis =1 )]

outlier_dict = {}
for col in sc_df_0.columns:
  outlier = sc_df_0[col][sc_outlier_df[col]]
  outlier_1 = outlier.reset_index().values.tolist()
  outlier_2 = list(set([l[-1] for l in outlier_1]))
  outlier_dict.update({col:outlier_2})
print(outlier_dict)

outlier_df = pd.DataFrame(outlier_dict)
```
